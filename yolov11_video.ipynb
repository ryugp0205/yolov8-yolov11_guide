```python
# =========================================
# 1Ô∏è‚É£ ÌïÑÏàò ÎùºÏù¥Î∏åÎü¨Î¶¨ ÏÑ§Ïπò
# =========================================
!apt -y install ffmpeg
!pip install -U yt-dlp
!$(which python3) -m pip install -U ultralytics

# =========================================
# 2Ô∏è‚É£ Ïú†ÌäúÎ∏å ÏáºÏ∏† ÏòÅÏÉÅ Îã§Ïö¥Î°úÎìú
# =========================================
VIDEO_URL = "https://www.youtube.com/shorts/sF02GVXE7Sc"
!yt-dlp -f best -o "video.mp4" {VIDEO_URL}

# Îã§Ïö¥Î°úÎìú ÌôïÏù∏
import os
if not os.path.exists("video.mp4"):
    raise FileNotFoundError("‚ùå ÏòÅÏÉÅ Îã§Ïö¥Î°úÎìú Ïã§Ìå®! Ïú†ÌäúÎ∏å ÎßÅÌÅ¨Î•º ÌôïÏù∏ÌïòÏÑ∏Ïöî.")
else:
    print("‚úÖ ÏòÅÏÉÅ Îã§Ïö¥Î°úÎìú ÏôÑÎ£å!")

# =========================================
# 3Ô∏è‚É£ YOLO Î™®Îç∏ Î°úÎìú Î∞è Ï∂îÏ†Å Ïã§Ìñâ
# =========================================
from ultralytics import YOLO
from IPython.display import Video, display

# Î™®Îç∏ Î°úÎìú (ÏûêÎèô Îã§Ïö¥Î°úÎìú)
model = YOLO("yolo11n.pt")

# Ï∂úÎ†• Ìè¥Îçî ÏÉùÏÑ±
os.makedirs("output", exist_ok=True)

# Í∞ùÏ≤¥ Ï∂îÏ†Å ÏàòÌñâ (Í≤∞Í≥º ÏûêÎèô Ï†ÄÏû•)
print("üöÄ Í∞ùÏ≤¥ Ï∂îÏ†Å Ï§ë... Ïû†ÏãúÎßå Í∏∞Îã§Î†§Ï£ºÏÑ∏Ïöî.")
results = model.track(source="video.mp4", save=True, save_dir="output", conf=0.3)

# =========================================
# 4Ô∏è‚É£ Í≤∞Í≥º ÏòÅÏÉÅ ÌååÏùº Ï∞æÍ∏∞
# =========================================
output_path = None
for root, dirs, files in os.walk("output"):
    for file in files:
        if file.endswith(".mp4"):
            output_path = os.path.join(root, file)

# =========================================
# 5Ô∏è‚É£ ColabÏóêÏÑú Í≤∞Í≥º ÏòÅÏÉÅ Ïû¨ÏÉù
# =========================================
if output_path:
    print(f"‚úÖ Í≤∞Í≥º ÏòÅÏÉÅ Ï†ÄÏû• ÏôÑÎ£å: {output_path}")
    display(Video(output_path, embed=True, width=640, height=360))
else:
    print("‚ùå Í≤∞Í≥º ÏòÅÏÉÅÏù¥ ÏÉùÏÑ±ÎêòÏßÄ ÏïäÏïòÏäµÎãàÎã§.")
```
